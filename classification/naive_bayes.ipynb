LOAD DATASET

```python
import seaborn as sns

data = sns.load_dataset("titanic")
data
```

DATA CLEANING

```python
data.drop(['sibsp','parch','alive','deck','embarked','who','embark_town','adult_male'],axis='columns',inplace=True)
data.head()
```

```python
import pandas as pd

data = pd.get_dummies(data,columns=['sex'])
```

```python
data
```

LABEL ENCODING FOR CLASS COLUMN

```python
data['class'].value_counts()
```

```python
data['class']= data['class'].map({'Third':3,'Second':2,'First':1})
data['alone']= data['alone'].map({True:1,False:0})
data['sex_male']= data['sex_male'].map({True:1,False:0})
data['sex_female']= data['sex_female'].map({True:1,False:0})
```

```python
data.head()
```

```python
data.info()
```

HANDLING NULL VALUES OF AGE COLUMN

```python
data['age'].mean()
```

```python
data.fillna({'age':data['age'].mean()},inplace=True)
data.info()
```

```python
target = data['survived']
input = data.drop(columns='survived')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(input,target,test_size=0.2)
```

```python
print(len(X_train))
print(len(X_test))
```

```python
from sklearn.naive_bayes import GaussianNB
model =GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```
